### Ocupación

```python
ocupacion = pd.read_csv('Ocupacion.txt', header = 1, sep='|', 
                        names=["OFICINA","CORRESPON","FECHA","HORA","GRUPO"], 
                        encoding='latin-1', 
                        dtype={'OFICINA':str, 'CORRESPON':str, 'FECHA':str, 'HORA':int, 'GRUPO':str},
                        parse_dates= ['FECHA'])
#ocupacion.dtypes
```

```python
ocupacion['DIA']=ocupacion['FECHA'].dt.weekday_name
#ocu = pd.DataFrame(ocupacion.groupby('GRUPO').count())
ocupacion['OCUALTA']= np.where(ocupacion['GRUPO']=='ALTO', 1, 0)
ocupacion['OCUMEDIA']= np.where(ocupacion['GRUPO']=='MEDIO', 1, 0)
ocupacion['OCUBAJA']= np.where(ocupacion['GRUPO']=='BAJO', 1, 0)
```

```python
ocupacion.head(20)
```

```python
#Puntuo la ocupacion, donde 1 es BAJA y 0 es ALTA y saco el promedio por dia y hora. Mayor puntuación, más recomendado.
ocupacion['OCUP_INDICE']= np.where(ocupacion['GRUPO']=='ALTO', 0, np.where(ocupacion['GRUPO']=='MEDIO', 0.5, 1))
ocupacion_ddhh = pivot_table(ocupacion, values='OCUP_INDICE', index=["OFICINA","CORRESPON"],columns=['DIA','HORA'], aggfunc=np.average)
ocupacion_ddhh.columns =['_'.join(str(s).strip() for s in col if s) for col in ocupacion_ddhh.columns]
ocupacion_ddhh.head(20)
```

#### Oficinas + Ocupación

```python
oficinas_ocupacion=pd.merge(oficinas, ocupacion_ddhh, on=["OFICINA","CORRESPON"], how='left')
oficinas_ocupacion.insert (0, 'TIPO',"OFICINA")
oficinas_ocupacion.head(20)
```

```python
cnt = len(oficinas_ocupacion)
print(cnt)
```

### Cajeros

```python
cajeros = pd.read_csv('Cajeros.txt', header = 0, sep='|', 
                      names=["CAJERO","GRUPO_CAJAMAR","ENTIDAD","OFICINA","NOMBRE_CALLE","NUMERO","COD_POSTAL","POBLACION",
                             "PROVINCIA_OLD","LATITUD","LONGITUD","LIBRETA","INGRESO","RECIBOS","CONTACTLESS","TRANS",
                             "EXTRAER","CON_SALDO","COSTE_EXTRACCION","EXCEPCIONES"],
                      na_values ="NA", 
                      dtype={'CAJERO':str, 'ENTIDAD':str,"OFICINA":str, 'NUMERO':str, 'COD_POSTAL':str, 
                             'PROVINCIA':str},
                      encoding='latin-1')
#cajeros.dtypes
```

```python
cnt = len(cajeros)
print(cnt) 
```

```python
#Hay oficinas con NA
#cajeros_filter = cajeros[cajeros['OFICINA'].isna()]
#cajeros_filter.head(100)

#relleno con 9999 las oficinas NA
cajeros["OFICINA"] = cajeros["OFICINA"].fillna("9999")
```

```python
#corrijo el formato agregando 0 a la izquierda
cajeros["OFICINA"] = cajeros.OFICINA.str.rjust(4, '0')
#cajeros.head(20) 
```

```python
# Si el NUMERO es 0 lo pongo vacío
cajeros.replace({'NUMERO' : '0'}, '', inplace=True)
#cajeros.head(20) 
```

```python
# Junto Nombre y Número de la calle en Campo Dirección 
cajeros.insert(4, 'DIRECCION',cajeros.NOMBRE_CALLE.astype(str).str.cat(cajeros.NUMERO.astype(str), sep=' '))
```

```python
# Elimino NOMBRE_CALLE y NUMERO
cajeros = cajeros.drop(['NOMBRE_CALLE','NUMERO'], 1) 
```

```python
# Formateo variable Grupo Cooperativo Cajamar para visualización
cajeros.insert(1, 'GRUPO_CAJAMAR_SN',np.where(cajeros['GRUPO_CAJAMAR']==1,'S', 'N'))
```

```python
#provincias_caj = pd.DataFrame(cajeros.groupby('PROVINCIA_OLD').count().CAJERO)
#provincias_caj.head(20) 
```

```python
# Corrijo nombres Provincias y agrego Código Provincia
#cajeros=cajeros.join(prov_aux, lsuffix='PROVINCIA_OLD', rsuffix='PROVINCIA_OLD')
cajeros=pd.merge(cajeros, prov_aux, on='PROVINCIA_OLD', how='left')
cajeros = cajeros.drop(['PROVINCIA_OLD'], 1) 
```

### Estados

```python
estados = pd.read_csv('Estados.txt', header = 1, sep='|', 
                      names=["CAJERO","FECHA","HORA","ESTADO"], 
                      encoding='latin-1',
                      dtype={'CAJERO':str, 'FECHA':str, 'HORA':int, 'ESTADO':int}, 
                      parse_dates= ['FECHA'])
#estados.dtypes
```

```python
estados['DIA']=estados['FECHA'].dt.weekday_name
#ocu = pd.DataFrame(ocupacion.groupby('GRUPO').count())
estados['OPERATIVO']= np.where(estados['ESTADO']==0, 1, 0)
estados['OFFLINE']= np.where(estados['ESTADO']==1, 1, 0)
```

```python
#estados_filter = estados[estados['ESTADO']==0]
#estados_filter.head(200)
```

```python
#Calculo el promedio de Operativo. Mayor promedio, mas operativo, mas recomiendo.
estados_ddhh = pivot_table(estados, values='OPERATIVO', index=['CAJERO'],columns=['DIA','HORA'], aggfunc=np.average)
estados_ddhh.columns =['_'.join(str(s).strip() for s in col if s) for col in estados_ddhh.columns]
estados_ddhh.head(20)
```
